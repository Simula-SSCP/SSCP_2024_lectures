# Modeling reaction rates with the Law of Mass Action

We start with looking at how to model simple chemical reactions. To do this we will use the *law of mass action*, which is one of the most fundamental concepts in chemistry. Note however that the mathematical models we define here are more general than chemical reactions themselves, and can also be applied to other processes in the cell.

As an example, we want to look at a *binding reaction*, between oxygen (O$_2$) and myoglobin (Mb). Myoglobin is a oxygen-binding globular protein found in muscle cells. If an environment contains both free oxygen molecules and myoglobin, the two will spontaneously bind together to form a *oxymyoglobin* complex (MbO$_2$).

<center><img src="fig/myoglobin_reaction.png" width=600></img></center>

<center><div style="max-width:750px; word-wrap:break-word;"><b>Figure 1:</b> Myoglobin is a complex globular protein, but for our example, the only important thing is that it has an oxygen binding group at its center which can bind a single oxygen protein. <br>(Image adapted from <a href="https://en.wikipedia.org/wiki/File:PicketFenceGenericRevised.png">Wikimedia</a> and used under CC BY-SA 3.0). </br></div></center>

This can be written as the reaction equation

\begin{equation}
\mathrm{Mb} + \mathrm{O_2} \overset{k}{\longrightarrow} \mathrm{MbO_2}.
\end{equation}

This reaction occurs *spontaneously* (i.e., on its own), but not *instantaneously*. Instead it will progress at a certain *rate*, and this is where the law of mass action comes in.



#### What decides the rate of a reaction?

For a binding reaction to occur at all, the two reacting molecules need to be close enough to each other to interact. At the molecular level, the reactants are not stationary, but move around in a random and chaotic manner called thermal motion. Due to this motion, many meetings between the two reactants will occur. Not every such meetings will result in a binding. Some will, and some will not. But it is clear that the number of bindings that occur in *total* should be proportional to the number of such meetings that occur. Because thermal motion is inherently random, it is very reasonable to assume that the number of meetings will be proportional to the *concentrations* of the reactants. The more molecules you have available, the more meetings will occur. If we use square brackets to denote the concentrations of the reactants, we write

\begin{equation}
\text{rate} \propto [\mathrm{Mb}]\times[\mathrm{O_2}].
\end{equation}

Thus the *rate* of the binding reaction is proportional to the product of the concentration of the reactants. If you are not familiar with it, the $\propto$ symbol means "proportional to". Instead of writing out the word "rate" in our mathematical equations, we typically write it out as the derivative of the concentrations themselves. If the rate is high, concentrations will change quickly, meaning a high derivative. So we have

\begin{equation}
\frac{\mathrm{d[MbO_2]}}{\mathrm{d}t} \propto [\mathrm{Mb}]\times[\mathrm{O_2}].
\end{equation}


If you found the whole explanation of thermal motion and molecular meetings to be confusing - ignore the finer details for now. The main point is that the rate of *any* reaction is proportional to the concentrations of the reactants themselves. And that is precisely what the law of mass action states.

#### Writing out a set of differential equations

To find an actual equation we can solve, we need an actual equality, not just a proportionality. We therefore introduce the proportionality constant $k$, which is typically called the *rate constant*. Thus we have the three equations

\begin{equation}
\frac{\mathrm{d[MbO_2]}}{\mathrm{d}t} = k\mathrm{[Mb][O_2]}, \qquad \frac{\mathrm{d [Mb]}}{\mathrm{d}t} = -k\mathrm{[Mb]}\mathrm{[O_2]}, \qquad \frac{\mathrm{d [O_2]}}{\mathrm{d}t} = -k\mathrm{[Mb]}\mathrm{[O_2]}.
\end{equation}

Note that we get three equations, because we have three concentrations that are changing. If binding occurs, we get more oxymyoglobin, but less free oxygen and myoglobin. These three equations are now effectively a set of ordinary equations we can solve to find how the concentrations change over time.

### Solving the ODE system

We have now defined a system of three ODEs we can solve to find how the concentrations of three species change over time. This is an example of what we call an *initival value problem* in mathematics. Let us now take the time to solve this system numerically using Python. To do this we jump over to another notebook:

#### The rate constant

While we call $k$ the rate *constant*, note that it might not necessarily be all that constant. It is related to the probability of a reaction occurring, and so will be different for different chemical reactions. They will also depend on things like pH, temperature, pressure and so on. If factors like these change during a chemical reaction, the rate constants might very well also change. The name "constant" is meant in relation to *concentrations*, as the rate constant is independent of the concentrations themselves. As concentrations change, the *rate* will change, but the *rate constant* will not.

Note also that while we state that the rate is proportional to the concentrations of the reactants, it would be more accurate to say it is proportional to their *activities*. The activity of a given molecular species can be described as the *effective concentration* of a species and it is generally lower than the actual concentration in a more complex solution that do not behave like *ideal solutions* were the reactants are free to move around independently of each other. The details are not that important for our current discussion. But if a chemist read these lecture notes, they would probably scold me for being so careless with my choice of words.

#### Determining and scaling rate constants

To actually apply the law of mass action and model concentrations of a system over time, we will need to know the rate constants of any applicable reaction. These should therefore be considered *model parameters*. Put very simply, rate constants must typically be found empirically, or they must be adjusted to make a given model fit empirical data.

If we do not have empirical data available to use, of course most known chemical reactions have published data that we can lean on. It is however important to remember that rate constants most likely depend on things like temperature. Rate constants found under standard lab settings ($\approx 20^\circ$C), will need to be adjusted to be applicable in a physiological setting ($\approx 37^\circ$C). There are techniques for scaling rate constants in this way, typically done by using a so-called temperature coefficient, denoted $Q_{\rm 10}$, which tells you how much a given model parameter needs to be scaled. But let us ignore such details for now. You can read more about it here if you are curious:
* https://en.wikipedia.org/wiki/Q10_(temperature_coefficient)


#### Activation energy and the Arrhenius equation

To get a better understanding of rate constants, and why chemical reactions typically speed up with temperature, we can turn to the *Arrhenius Equation*. This simple equation states that a reaction rate typically can be written out on the form

\begin{equation}
k = Ae^{-E_{\rm a}/RT}.
\end{equation}

Here $k$ is a rate constant, $A$ is called the pre-exponential factor which is dependent on what reaction we are looking at, but independent of temperature. Then we have an exponential term where $E_{\rm a}$ represents the *activation energy* of the reaction, $R$ the ideal gas constant and $T$ the temperature (in absolute units).

The activation energy $E_{\rm a}$ represents the energy the reactants need to overcome to actually react. They need enough energi to overcome a "barrier" and undergo a chemical shift to change their structures. The bigger the barrier, the harder it is to react, and the slower the rate of the reaction at the macroscopic level.

Arrhenius' equation tells us that the rate constant has an exponential dependence on temperature, and as the temperature is lowered, the rate constant is dampened. When the temperature increases, the thermal motion of the reactants is energized and so it will be easier to overcome the activation barrier and actually react.

## Reversible reactions

When formulating our model for the binding reaction, we made an implicit assumption that the *reverse* reaction doesn't occur, i.e., an oxymyoglobin splitting into an oxygen molecule and a myoglobin, a process we call *dissociation*. Many chemical reactions are reversible, and will occur in both directions, especially so for a closed system such as the inside of a cell. The binding and dissociation of oxygen will occur in parallel and we therefore modify our chemical equation to reflect this

\begin{equation}
\mathrm{Mb} + \mathrm{O_2} \underset{k_-}{\overset{k_+}{\rightleftharpoons}} \mathrm{MbO_2}.
\end{equation}
Here we have replaced the unidirectional arrow with a bi-directional pair of arrows representing both the forward and reverse directions of the reaction. The two directions of a reaction do not necessarily have the same rate constant, and so we denote the two seperate rate constants $k_+$ and $k_-$ to tell them apart.

The rate of the forward reaction will be the same as before, and the rate of the reverse reaction can be formulated as

\begin{align}
\text{forward rate} &= k_+\mathrm{[Mb]}\mathrm{[O_2]}, \\
\text{reverse rate} &= k_-\mathrm{[MbO_2]}.
\end{align}

The rate is again found by assuming the rate to be proportional to the reactants concentrations, and this is true even if the reactant is a single molecule. This makes sense, as any given oxymyoglobin molecule in the system will have a given probability of dissociating at any given time, meaning the total amount of molecules splitting up per time should be proportional to the number of oxyglobin, i.e., their concentration.

In an environment containing both oxygen and myoglobin we will now have a complex dynamic situation where myoglobin and oyxgen are continuously binding together and dissociating. The *net* change in concentrations over time can be found by looking at the difference between the forward and reverse rate. The ODE describing the concentration of oxymyoglobin will for example be given  by

If we now want to formulate the change in oxymgoblin over time, 

\begin{equation}
\frac{\mathrm{d[MbO}_2]}{\mathrm{d}t} = k_+[\mathrm{Mb}][\mathrm{O_2}] - k_-[\mathrm{MbO_2}].
\end{equation}

The first term corresponds to the forward rate and is positive, because oxymyoglobin is formed when a binding reaction occurs. The second term is the reverse rate, which we subtract because oxymyoglobin are removed from the system when they dissociate. 

## Chemical Equilibrium

When modeling a unidirectional chemical reaction, the concentration of the reactants can only drop over time while the concentrations of the product can only increase. When we look at a reversible reaction however, the concentrations of the different molecules can both increase and decrease. Looking at ODE we just formulated for example, we see the concentration of oxymyoglobin increases if the forward rate is larger than the reverse, and decreases if the situation is flipped. As the concentration changes over time however, these two rates will also change and approach each other until we end up in a chemical *equilibrium*.

An equilibrium is reached once the forward and reverse rate of the reaction are of equal magnitude and effectively cancel each other out. We can write this as the following equilibrium criteria

\begin{equation}
\frac{\mathrm{d[MbO}_2]}{\mathrm{d} t} = k_+[\mathrm{Mb}][\mathrm{O_2}] - k_-[\mathrm{MbO_2}] = 0 \quad (\text{at equilibrium}).
\end{equation}

This criteria does not say that the reaction has stopped, but rather that the forward and reverse reactions are now occurring at the same rate, meaning there is no *net* change in concentrations. This is referred to as a *dynamic* equilibrium, to emphasize that the reaction is still occurring at the molecular level.

The equilibrium condition states that the two rates are equal

\begin{equation}
    k_+[\mathrm{Mb}][\mathrm{O_2}] = k_-[\mathrm{MbO_2}] \quad (\text{at equilibrium}).
\end{equation}

And from this criteria we can see that the *ratio* of the reactants and the product must be constant at equilibrium

\begin{equation}
\label{eq:equilibrium_equation}
\frac{[\mathrm{Mb}][\mathrm{O_2}]}{[\mathrm{MbO_2}]} = \frac{k_-}{k_+} \equiv K_{\rm d} \quad (\text{at equilibrium}).
\end{equation}

The ratio of the rate constants is typically referred to as the *dissociation constant* of the reaction, and it is denoted $K_{\rm d}$. The higher the dissociation constant, the higher the concentrations of the free molecules will tend to be at equilibrium. Some refer to the dissociation constant simply as the *equilibrium* constant, $K_{\rm eq}$. Note however that the term equilibrium doesn't indicate which direction of the reaction we are referring and so this term could also mean the inverse relation $k_+/k_-$, which would be the *association* constant. There are no strict conventions of which of these are used, and some care should be taken to interpret how the equilibrium constant for a reaction is formulated and used.

This equation is referred to as *the equilibrium equation* or *the equilibrium condition* and it describes how the ratio of concentrations will settle at equilibrium. It is a direct consequence of the law of mass action. If one is referring to the law of mass action as an equation, it is typically the equilibrium equation that is meant. Note that the equilibrium equation does not hold at all times, but only when a system is at equilibrium. However, any chemical system will spontaneously tend towards this state, and for many reactions this occurs so swiftly that one can consider the system to always be at equilibrium for all practical purposes (typically called a *quasistatic* process). 

The equilibrium equation is very useful, but it might be a bit challenging to see how this equation is interpreted or used in practice. We will shortly look at an illustrative example, but let us first take the time to solve the system of ODEs for the reversible reaction and also verify that the system does indeed tend towards equilibrum.

### Fractional binding

So far we have treated the binding reaction of myoglobin and oxygen to be occuring in a closed system, where the total concentrations of all three species is constant over time. However, in a muscle cell, this might not actually be the most realistic assumption. Inside the cell we can typically think of the total amount of *myoglobin* as fixed, i.e.,

$$[\mathrm{Mb}]_{\rm tot} = [\mathrm{Mb}] + [\mathrm{MbO_2}] = \mathrm{const.}$$

The total amount of oxygen in the cell however, can change over time. Oxygen is supplied by the circulatory system and is free to diffuse into and out of the cell. When oxygen binds to myoglobin, the amount of free oxygen inside the cell will decrease, and so more oxygen will tend to spontaneously enter the cell. This means that the total amount of oxygen in the cell is *not* constant. Instead we can assume that it is the concentration of free oxygen $[\mathrm{O}_2]$, that is constant within the cell. At least over shorter time scales.

We now introduce a new quantity, the *fractional degree of oxygenation*, i.e., the fraction of all myoglobin molecules that will have oxygen bound to them. This corresponds to the following fraction

\begin{equation}
Y = \frac{[\text{bound Mb}]}{[\text{total Mb}]} = \frac{[\mathrm{MbO_2}]}{[\mathrm{Mb}] + [\mathrm{MbO_2}]}.
\end{equation}

We will now use the equilibrium equation to find an expression of $Y$ as a functional of the free oxygen concentration. To simplify this derivation, we are only concerned with the fractional oxygenation at equilibrium. We start by solving the equilibrium equation for $[\mathrm{MbO_2}]$, giving

\begin{equation}
[\mathrm{MbO_2}] = K_{\rm d}^{-1}[\mathrm{Mb}][\mathrm{O_2}].
\end{equation}

We can now insert this into our expression for $Y$, yielding

\begin{equation}
Y = \frac{K_{\rm d}^{-1}[\mathrm{Mb}][\mathrm{O_2}]}{[\mathrm{Mb}] + K_{\rm d}^{-1}[\mathrm{Mb}][\mathrm{O_2}]}.
\end{equation}

Simplifying, we find a much neater expression:

\begin{equation}
\label{eq:myoglobin}
Y = \frac{[\mathrm{O_2}]}{[\mathrm{O_2}] + K_{\rm d}}.
\end{equation}

To summarize, this expression describes what ratio of myoglobin molecules will have oxygen bound to them as a function of the concentration of free oxygen. We graph this expression in the following figure.

<center><img src="fig/myoglobin.png" width=500></center>

<center><div style="max-width:600px; word-wrap:break-word;"><b>Figure 3:</b> The fractional degree of oxygenation of myoglobin as a function of free oxygen, which we have scaled by the dissociation constant $K_{\rm d}$.</div></center> 


It is worth taking a few minutes to think about the implications of this curve. For one, we have effectively assumed that the amount of oxygen available is an endless resource, so one *might* assume that this would imply over time that all myoglobin molecules would end up saturated with oxygen. But from the binding curve we see that this is *not* the case, even when the oxygen concentrations grow very large, there will be some small ratio of myoglobin that are not oxygenated. This is simply because of rates of the reaction. If all myoglobin were oxyganted, the forward rate of the binding reaction would be zero, but the reverse will be very high - and so it is clear that fractional degree of oxygenation will stabalize at some number below 100%.

The curve tells us what number this will be. If there is no oxygen available, no myoglobin can bind oxygen either. As some oxygen is added to the system some myoglobin will start binding it. If the concentration of free oxygen is low, the fractional oxygenation will also be low, regardless of how much oxygen is available *in total*, it is the concentration of oxygen that is important, due to the law of mass action. As the concentration of free oxygen increases, so does the degree of oxygenation. Note especially that when $[{\rm O}_2] = K_{\rm d}$, we get $Y=0.5$. In this scenario, the dissociation constant is equal to the point of half-saturation of the system. As the oxygen levels rise higher than this the oxygenation rises slower and slower, and you can see the oxygen concentrations need to be very high for the oxygenation of myoglobin to approach full saturation.

In this example we have derived the fractional oxygenation of myoglobin using the law of mass action and the equilibrium equation. This might seem like a very specific example, especially due to our assumption of a constant level of free oxygen. However, all derivations we have done come directly from the law of mass action and are therefore very general. The assumption of a constant amount of free oxygen can also be a good approximation of a system where there is plentiful of one reactant, and comparatively less of the other. At the end of this session we will turn to enzyme kinematics, which will be precisely such a system, with a lot of a reactant compared to a small amount of a given enzyme.

## Modeling Cooperativity and The Hill Equation

We now turn to an important phenomenon in biophysical systems known as *cooperativity*, which in turn will lead us to the *Hill equation*. The example we use to derive the Hill equation will mirror the derivation we just carried out, and will again rely directly on the law of mass action.

We now introduce a different oxygen-binding protein called *hemoglobin*. A myoglobin protein can bind a single oxygen molecule, but a hemoglobin can bind *up to four* oxygen molecules at once. Put simply, a hemoglobin molecule is a lot like four individual myoglobin proteins stapled together.

To model how hemoglobin binds oxygen we can envision each hemoglobin to be in one of five states, ranging from having zero oxygen molecules bound up to having all four bonding sites occupied. As illustrated in the following figure

<center><img src="fig/hemoglobin_reaction.png" width=600></center>

<center><div style="max-width:600px; word-wrap:break-word;"><b>Figure 4:</b>  A single hemoglobin molecule has four binding sites for oxygen.</div></center> 

To properly model this system we would need to set up six ODEs, one for each species in the system (five different versions of Hb, and one for free oxygen). However, we can simplify the system considerably. If it were the case that each of these binding and dissociation reactions had the same rate constants, we could simplify this system considerably and say that it behaves exactly like myoglobin, but at four times the concentration. In reality however, the rate constants are far from constant, instead hemoglobin has the interesting property that it binds oxygen more readily the more oxygen it already has bound. This is known as *cooperative binding*, as the binding sites seem to be cooperating with each other.

The net results of cooperative binding is that hemoglobin will tend to be found at either extremes, either with no oxygen bound, or all four oxygen bound, i.e., an all-or-nothing binding. To model this binding let us make the simplifying assumption that hemoglobin binds all four oxygen molecules simultaneously. This assumption isn't strictly true, but it is a very useful approximation. 

<center><img src="fig/hemoglobin_reaction_simplified.png" width=200></center>

<center><div style="max-width:600px; word-wrap:break-word;"><b>Figure 4b:</b> Due to cooperative binding, hemoglobin tends to bind oxygen in an all-or-nothing fashion. We therefore approximate the reaction as a single step.</div></center> 

Under this approximation the reaction simplifies to a single reversible reaction, described by the following chemical equation

\begin{equation}
\mathrm{Hb} + 4\mathrm{O_2} \underset{k_-}{\overset{k_+}{\rightleftharpoons}} \mathrm{Hb(O_2)_4}.
\end{equation}

We can again apply the law of mass action to find the rate of this reaction. Because four oxygen molecules are now to bind simultaneously, the rate will be proportional to the concentration of all five reactants, meaning the reaction rate will be proportional to the oxygen concentration to the fourth power

\begin{equation}
\text{forward rate} \propto [\mathrm{Hb}] \times [\mathrm{O}_2] \times [\mathrm{O}_2] \times [\mathrm{O}_2] \times [\mathrm{O}_2] = [\mathrm{Hb}][\mathrm{O}_2]^4.
\end{equation}

Put simply, having the reaction depend on four oxygen molecules makes the rate dependence on the oxygen concentration a lot steeper. Writing out the full forward and reverse rates gives the change in oxygen-bound hemoglobin to be 

\begin{equation}
\frac{\mathrm{d[Hb(O}_2)_4]}{\mathrm{d} t} = k_+ \mathrm{[Hb]}\mathrm{[O_2]}^4 - k_- [\mathrm{Hb(O}_2)_4].
\end{equation}

Which in turn yields the same equilibrium equation as earlier, but with the oxygen concentration taken to the fourth power

\begin{equation}
\frac{\mathrm{[Hb]}\mathrm{[O_2]}^4}{\mathrm{[Hb(O_2)_4]}} = K_{\rm d}.
\end{equation}

The derivation is so far very similar to the one we made for myoglobin, and the only difference is in the power of four. However, it is no immediately apparent how this exponent affects the system in practical terms. Let us therefore derive the fractional degree of oxygenation of hemoglobin, so we can compare this to the one we found for myoglobin. Finding this expression is practically identical to how we did it earlier, and it can be a good exercise for the reader to carry this out. You should find the following function

\begin{equation}
Y = \frac{\mathrm{[O_2]}^4}{\mathrm{[O_2]}^4 + K_d}.
\end{equation}

### Comparing the fractional binding curves of myoglobin and hemoglobin

If you compare the expressions we found for the fractional degree of oxygenation for myoglobin and hemoglobin, the only difference is the exponent on the concentration of free oxygen. Both equations are on the form 

\begin{equation}
Y = \frac{\mathrm{[O_2]}^n}{\mathrm{[O_2]}^n + K_d},
\end{equation}

This equation is a specific example of the *Hill equation*, and the parameter $n$ is typically called the Hill coefficient. Thus, oxygen binding to myoglobin has a hill coefficient of $n=1$, while for hemoglobin it is $n=4$.

Note also that some care should be taken when discussing the dissociation constant $K_{\rm d}$. In our case, we introduced it as $k_{\rm -}/k_{\rm +}$, but for our expression of $Y$ to make sense, notice that $K_{\rm d}$ has to have units of mol$^{n}$. For myoglobin this is simply units of concentration, but for hemoglobin it is a lot less intuitive. Some therefore prefer to define a quantity $K_{\rm A}$, which obeys

\begin{equation}
(K_{\rm A})^n = K_{\rm d}.
\end{equation}

This quantity is usually referred to as the *microscopic* dissociation constant, while $K_{\rm d}$ can be referred to as the *apparent* dissociation constant. From the definition $K_{\rm A}$ will have units of concentration and it is when $[\mathrm{O}_2] = K_{\rm A}$ that the binding will be half-saturated.

Again, there are no strict conventions across the literature, and some refer to the microscopic dissociation constant as $K_{\rm d}$, which can definitely lead to confusion. Because of the varying conventions, we recommend you be extra mindful of units when modeling chemical reactions and ensure you are using the appropriate parameters.

To better understand the impact of the difference in the Hill coefficient between the two functions, let us plot the two curves in the same figure, for the same half-saturation point $K_{\rm A}$. This graph is shown below. We see that by going from a Hill coefficient of 1 to 4 makes the function a sigmoidal S-shape with a much steeper rise. The two curves intersect in the half-saturation point (because they have the same $K_{\rm A}$), but the curve for $n=4$ goes towards zero much faster below this point, and goes to one much faster above it.

When comparing myoglobin and hemoglobin however, it would be misleading to use this figure, because the two have very different dissociation constants as well. In the lungs, the partial pressure of oxygen is on the order of 100 mm Hg, while in muscle tissue, it is about 20 mm Hg. If we compare the fractional binding curves of myoglobin and hemoglobin at these two levels, we see that the two reach about the same saturation in the lungs, but in the muscles, hemoglobin has a much lower fractional binding. Effectively, this means that hemoglobin is much better suited to act as an oxygen carrier, as it can pick up oxygen in the lungs, and deposit this in the muscles. Myoglobin on the other hand, does not act that well as an oxygen carrier, because it struggles to effectively part with its oxygen unless there is very little oxygen present.

The difference between myoglobin and hemoglobin make them suited for different tasks. Hemoglobin is found primarily in the blood, where it acts as an oxygen carrier, while myoglobin is found in tissue where it acts more like an oxygen reserve or storage. This functional difference between the two is a results of the cooperativity of the four hemoglobin subunits. This is a common theme in biological systems, where cooperativity between different units will make transitions steeper and functionally different. The Hill equation is therefore very general, and is useful for far more than modeling oxygen-binding.






### The general Hill equation

The Hill equation was derived in 1910 by Archibald Hill specifically to describe the binding curve of hemoglobin. The equation is however much more general than this specific example, and can be used to model many types of binding reactions. To emphasize this, let us write it out as a more general equation. In biochemistry, it is common to call the ion or molecule that is being bound the *ligand* and the macromolecule that binds it the *receptor*. A general ligand-receptor binding can then be described with the equation

\begin{equation}
\theta = \frac{[L]^n}{[L]^n + K_{\rm d}}.
\end{equation}

Here, $\theta$ is the fractional binding, $[L]$ the free ligand-concentration, $K_{\rm d}$ the dissociation constant of the binding reaction and $n$ the Hill coefficient.

### Interpreting the Hill coefficient

In our example, we showed that the binding curve of hemoglobin has a Hill coefficient of 4, because it binds four oxygen molecules in an all-or-nothing fashion. Recall however, that this was an approximation we made. If you were to measure the actual oxygen binding curve of hemoglobin you would get a result that fits very well to the Hill curve, but with a Hill coefficient of $n=3$ rather than 4. This is because hemoglobin can, and will, bind one, two or three oxygen molecules. Therefore, the Hill coefficient shouldn't be interpreted strictly as the number of units cooperating, but rather the *degree* of cooperativity. In fact, $n$ can even be fractional. Depending on the exact conditions under which you measure the oxygen binding, you'll find a best fit using a coefficient in the range 2.8-3.1. If we interpret the coefficient as a phenomenological parameter trying to catch the degree of cooperativity, using a fractional coefficient to get the best possible fit to measured data makes perfect sense.

In general, we can say that $n = 1$ means no cooperativity, i.e., independent binding sites. When we have 
cooperative binding, we have $n > 1$. Note also that one can have *negative cooperation*, meaning binding to one site inhibits binding to other sites. In such a case we would have a coefficient $n < 1$.

# Thermodynamics and Gibbs Free Energy

When analyzing or modeling any reaction or process occuring within the cell, it can be very useful to go back to the basics and consider the underlying physical constraints present. To do this we typically want to analyze the *thermodynamic quantities* of a system, namely energy and *entropy*. Such analysis can give us a better understanding of what processes will occur spontaneously, of how close they are to equilibrium and other physical limits. A very useful concept here is called *Gibbs Free Energy*, which combines energy and entropy into a single quantity.

Let us very briefly cover the fundamental laws of thermodynamics, then explain what Gibbs free energy is and how we can use it. 

#### First law of Thermodynamics

> Energy cannot be created or destroyed

The first law is a statement about conservation of energy. The total energy of any *isolated* system will always be constant over time. A part of the cell, or a reacting molecule within the cell, however, is not isolated and can exchange energy with its enviroment through either heat exchange or mechanical work. 

#### Second law of Thermodynamics

> In any spontaneous process, the total net entropy can never decrease, only increase

The second law of thermodynamics is conceptually much trickier than the first. And it can be stated in a lot of different, functionally equivalent, ways. 

Put simply, entropy is a measure of a systems *disorder* and it is a physical property much like energy. Like energy, the entropy of the reactants in a process does not necessarily have to increase, they *can decrease*, but the net entropy of the reactants and their environment combined can only increase. Because of this, the entropy of any isolated system will tend to grow over time and evolve towards some maximum possible entropy. This maximum will correspond to the systems thermodynamic equilibrium state. Put in another way, we can say that entropy is the driving force of chemical reactions and other cellular processes.

#### Gibbs Free Energy (Energy + Entropy combined)

The first and second laws of thermodynamics are physical properties of how nature works, they must always be fulfilled. Since the two regard different physical properties however, including both in any given analysis can be a bit tricky. To make this easier, we combine the two quantities into a single quantity, called the *Gibbs free energy*, or simply the *Gibbs energy* of the system. It can be expressed as

\begin{equation}
G = U + pV - TS.
\end{equation}

Here, $G$ is the Gibbs energy of the system, $U$, $V$ and $S$ are the systems internal energy, volume and entropy respectively. Lastly we have the pressure, $p$, and temperature, $T$, of the surrounding environment.

From this definition, $G$ will have units of energy, which is why we call it Gibbs *energy*, but note that entropy is also baked into this definition. With the introduction of Gibbs energy we can restate the second law of thermodynamics not in terms of entropy, but in terms of Gibbs energy. The net Gibbs energy of any isolated system cannot increase, but only decrease. An isolated system will spontaneously tend towards its minimal Gibbs energy, which will then correspond to its thermodynamic equilibrium.


### Why is it called Gibbs *free* energy?

Gibbs energy is known by many names, but many like to use the descriptor *free* energy. When J. W. Gibbs introduced the concept in the 1870s he dubbed it simply *available energy*. This is because the Gibbs energy can be thought of as the maximum amount of work that can be extracted as useful work from a given system. It would be impossible to extract more, without breaking the two laws of thermodynamics.

A good metaphor for Gibbs energy is found in Schroeder's *An Introduction to Thermal Physics*. Schroeder asks you to imagine a wizard conjuring up a rabbit from nothing. Obviously, the wizard has to respect the laws of thermodynamics, which state that energy cannot be created, and so he will have to supply the needed energy in the process. But how much energy would the wizard need to supply to create the rabbit? First he has to produce the internal energy of the rabbit itself, $U$, which encompasses all the mass energy and all internal chemical energy and so on. If the rabbit were to be created in a complete vacuum, this would be the end of it. However, to get the rabbit to pop into existence in front of you, the wizard actually has to make space for the rabbit by pushing some air aside, which costs mechanical energy. This cost would be $pV$, the higher the air pressure or the bigger the rabbit, the more work has to be performed. The energy bill is now $U + pV$. However, the wizard is quite clever, and so he can save some energy as well. If he conjures up a cold rabbit, he knows the laws of thermodynamics says heat will spontaneously flow from the high temperature surroundings into the low temperature rabbit, so the wizard can subtract this energy from his energy bill. This contribution is $TS$, where $T$ is the temperature of the surroundings and $S$ is the entropy of the rabbit. Why does the entropy enter into it? Because temperature is defined as the derivative $T = {\rm d}U/{\rm d}S$. So the more entropy the rabbit has, the more thermal energy it will absorb from its environment. The total energy bill the wizard needs to pay is therefore the Gibbs energy $U + pV - TS$.

<center><div style="max-width:600px; word-wrap:break-word;"><b>Figure 6:</b> Schroeder introduces a metaphor to explain the concept of Gibbs free energy. Envison a wizard conjuring a rabbit out of nothing. He would need to supply a total amount of energy equal to $G = U + pV - TS$.</div></center> 

Schroeder's analogy nicely illustrates how Gibbs energy combines different thermodynamics considerations into a single quantity. But if the Gibbs energy represents a *cost*, why is it called the *free* energy? The name describes the reverse process. If an evil wizard were to annihilate a rabit and extract as much energy as possible in the process, we would first get the internal energy $U$ and some extra mechanical work from the surrounding air collapsing into the resulting vacuum. However, they would have to dump som of that energy as heat into the environment to satisfy the second law. The Gibbs energy is thus the amount of energy a system could release if it were to cease existing.

Of course, in the real world, a system doesn't simply cease existing in this manner. However, any reaction or process can be thought of as one system being annihilated, and a new one being created. It is the *net change* in the Gibbs energy that is the important quantity.

### Change in Gibbs energy

In a given reaction, the energy, volume or entropy of the system can change. Any change in the pressure or temperature however, can be neglected, because we assume the surrounding environment to be much larger than the reacting system. Thus, the total change in Gibbs energy can generelly be written out as 

\begin{equation}
\Delta G = \Delta U + p\Delta V - T\Delta S.
\end{equation}

Thus the net change in Gibbs energy will be a weighted sum of its change in internal energy, $\Delta U$, its change in volume, $\Delta V$, or a change in net entropy $\Delta S$. 

Recall now that the net Gibbs energy of any system will tend to spontaneously decrease over time. So we can now state the following for any reaction or process:

* If $\Delta G < 0$, it can occur spontaneously
* If $\Delta G > 0$, it cannot happen (as this would break the first and/or second law)
* If $\Delta G = 0$, the system is at thermodynamic equilibrium.


Note that when we say any process where $\Delta G > 0$ cannot happen, we mean this for an isolated system. In certain reactions in the cell, one process with a $\Delta G > 0$ is *coupled* to another with $\Delta G < 0$, and as long as the *net* change is negative, this is fine as one process is used to power another.

#### Gibbs energy of molecular species

For any molecular species, we can express its Gibbs free energy as

\begin{equation}
G = G^0 + RT \ln c.
\end{equation}

Where the first term, $G^0$, is called the standard free energy, which is dependent on what chemical species is being discussed, but not the concentrations involved. The second term is dependent on the concentration of the species, but is otherwise inconsiderate of what the species is. Here, $R$ is the gas constant, $T$ the temperature and $c$ the concentration of the species.

If a reaction occurs where A $\to$ B, the net change in Gibbs energy will then be expressed as

\begin{equation}
\Delta G = G_{\rm B} - G_{\rm A} = (G_{\rm B}^0 - G_{\rm A}^0) + RT (\ln c_{\rm B} - \ln c_{\rm A}),
\end{equation}

which simplifies to

\begin{equation}
\Delta G = \Delta G^0 + RT \ln \frac{c_{\rm B}}{c_{\rm A}}.
\end{equation}

Note that the change in Gibbs energy can be written as two terms. The first is the change in standard free energy, which depends on what molecular species are interacting. This value is tabulated for most molecular species under standard conditions. The second term only cares about the ratio of the concentrations of the species, and so is not specific to what reaction is occurring.

#### Chemical Equilibrium from Gibbs energy

As statet, a system is in equilibrium with respect to a given reaction if the net change in Gibbs energy for that reaction is equal to zero. So is a system can inhabit to given states, A and B, then we know that at thermodynamic equilibrium we have

\begin{equation}
\Delta G = \Delta G^0 + RT \ln \frac{c_{\rm B}}{c_{\rm A}} = 0.
\end{equation}

If we take the exponential of this expression we find

\begin{equation}
\frac{c_{\rm B}}{c_{\rm A}} = e^{-\Delta G^0/RT}.
\end{equation}

The left hand side is now the ratio of the two concentrations, while the right hand side are all just constants. While this equilibrium condition looks very different to the one we derived when discussing the law of mass action, it effectively states the same thing. This also implies, perhaps not surprisingly, that the standard free energy $G_0$ for a given reaction is related to it's dissocation constant $K_{\rm d}$.

# Diffusion and Fick's law

The next physical principle that is useful to know about is *diffusion*. This is the process where concentration differences tend to smooth out over time. If a system has a non-uniform concentration of a given ion or molecule, then random thermal motion will tend to erase these differences over time. At the macroscopic level, this will look like there is a net *diffusive flux* of mass going against the concentration gradient. 

<center><img src="https://upload.wikimedia.org/wikipedia/commons/thumb/4/4d/DiffusionMicroMacro.gif/250px-DiffusionMicroMacro.gif" width=200></center>
<center><div style="max-width:600px; word-wrap:break-word;"><b>Figure 7:</b> A simple illustration of diffusion at the microscopic and macroscopic level taken from <a href="">Wikimedia Commons</a>. The top panel shows the thermal motion of a single particle, it is chaotic and random and essentially uniform in space. In the middle panel we see the motion of many such particles simultaneously. Because the initial distribution has more particles on the left than the right, more particles will tend to move right—even if they all are moving at random. In the bottom panel we get the macroscopic view, where the concentration evens out over time, quickly at first, then slower and slower.</div></center> 


To describe diffusion (at the macroscopic level) mathematically, we use Fick's law of diffusion (or Fick's *first* law of diffusion, as it is sometimes called). This law states that the *net* diffusive flux is proportional to the concentration gradient. In 1D it can be expressed as

\begin{equation}
J(x) = -D \frac{\partial c(x)}{\partial x},
\end{equation}

while in higher dimensions we would state it on vector form as

\begin{equation}
\vec{J}(\vec{r}) = -D \nabla c(\vec{r}).
\end{equation}

Here, $J$ is the net diffusive flux, $c$ is the concentration and $D$ the proportionality constant, called the *diffusivity* or *diffusion coefficient*. Note that while both $J$ and $c$ are field variables, meaning they have different values in different locations $\vec{r}$, we will from now on forgo writing their spatial arguments to get a neater notation. They are however, always spatially dependent.

Note that Fick's law holds true for each ionic or molecular species independently, so it is often written with a subindex denoting specific species, or generally as 

\begin{equation}
\vec{J}_k = -D_k \nabla c_k.
\end{equation}

where $k$ denotes the species. Note that the diffusivity is also species dependent.

Fick's law tells us that concentrations will spontaneously even out concentration differences. The bigger differences are present, the stronger diffusive flux we will have. In the next session we will start looking at a transport of ions across the cell membrane, and this transport is in large part driven by precisely such *diffusive flux*, because there are differences in ion concentrations inside and outside the cell.

## Enzyme Kinetics and the Michaelis-Menten equation

We now turn to a topic that can be described as an application of the law of mass action, which is enzyme kinematics. Enzymes are proteins that act as *cataclysts* for biochemical reactions, meaning they speed up reactions without being consumed in the process. 

There are many different physical or chemical mechanisms for how an enzyme can speed up a chemical process. We don't want to go into the fine details, but from a conceptual perspective we can simplify by saying that enzymes in some way lower the energy barrier of a chemical reaction. This can for example be achieved by binding to an intermediate state in the reaction. Recall now Arrhenius' law, which stated that a rate constant of a reaction could be written as
$$k = Ae^{-E_{\rm barrier}/k_{\rm B}T}.$$
Since the rate is proportional to the exponential of the activation energy, even a relatively small reduction in the barrier can cause a large increase in reaction speed. Because of this, enzymes are usually highly efficient and can speed up reactions by many orders of magnitude (as much as $10^7$ in some cases)!

The drastic increase in reaction speed caused by enzymes is hugely important for cellular function. For many reactions in the cell, their actual reaction rates are so slow that they are effectively stable. But as soon as a bit of enzyme is added, things start happening. In this sense, enzymes can be thought of more as on-buttons for various processes in the cell that are lying in stand-by. An example of this are the energy carrier molecules ATP. These molecules carry a fair bit of chemical energy, and will spontaneously dissocate and release their energy. Luckily, this spontaneous dissociation happens so slowly that ATP is stable enough to be an effective energy carrier. The cell can therefore produce ATP through metabolism and let it float around in the cell for a while, without fear of wasting energy. Throughout the cell are then molecular machines that are equipped with enzymes that speed up the dissocation of ATP, and harness the released energy in the process. Such enzymatic machines are known as the *ATPases*, and we will see several examples of them later in the course.

#### Enzymes change reaction speed, not direction

It is important to note that an enzyme can never change the net change in Gibb's energy in a reaction, and will therefore never be able to change the *direction* of a reaction. An enzyme cannot trigger a reaction that is not spontaneous on its own, it can just speed it up. The upside however, is that the enzyme molecules doesn't get used up in the reaction, and therefore even a small amount of enzyme is usually enough to trigger some process.

### Modeling enzymatic reaction

As mentioned, enzymes can work in many different ways. It can actually be challenging to explain precisely why they work. For example, the enzyme *catalase* is common to more or less all life that processes oxygen and is a very well studied organic molecule, but exactly how it works to break down H$_2$O$_2$ still isn't perfectly understood. Therefore, let us not try to understand *how* an enzyme works, but rather how it affects the reaction rate itself.

The model we will look at is called the Michaelis-Menten equation, and it is a direct result of the law of mass action—and in fact it closely resembles the Hill equation as well. While it is a very simple model, it is immensenly useful and sees a lot of use in physiological modeling. It sometimes goes under the name of *first-order kinetics*, as we make a few simplifying assumptions deriving it.

#### Model States

Enzymes generally work by binding to some molecule, the *substrate*, converting it, and then releasing the *product*. We therefore propose the following reaction to model this activity

$${\rm S} + {\rm E} \underset{k_{-1}}{\overset{k_1}{\rightleftharpoons}} {\rm ES} \overset{k_2}{\rightarrow} {\rm E} + {\rm P}.$$

Here S is the substrate that is to be converted and E is the enzyme, these can bind together and form an enzyme-substrate complex, ES. The ES complex can either break up again without converting the substrate, or the substrate can be converted and the complex breaks up into a free enzyme molecule and a product molecule, P. We assume that once the product has been released by the enzyme, it no longer binds to the enzyme spontaneously, so there is no reverse arrow in the final reaction. 

#### Reaction rate 

We are noe modeling a system with three different reactions, so what do we now mean by *reaction rate* in this case? The interesting quantity here is the rate at which the product $P$ is generated, i.e., the time derivative $\dot{[{\rm P}]}$. We will call this rate the 'reaction velocity' $v$. Luckily, the law of mass action applies to each concentration separately, so looking at the reaction equation, we can write this out as

$$v = \dot{[{\rm P}]} = k_2[{\rm ES}],$$

as you can see, to compute an actual value, we first need to find an expression for $[\rm ES]$. We start using the law of mass action for $\rm [E]$:

$$[\dot{{\rm E}}] = (k_{-1} + k_2)[{\rm ES}] - k_1[{\rm S}][{\rm E}].$$

To go further, we will make an assumption that the system will quickly reach an equilibrium where the ratio of bound and unbound enzyme is constant, and so $[{\rm E}]$ will be constant. If you recall, we mentioned that this is a common assumption that is typically called *quasistatic* conditions. From this assumption we we can set the derivative to zero to get

$$(k_{-1} + k_2)[{\rm ES}] = k_1[{\rm S}][{\rm E}].$$

The enzyme doesn't get "used up", so the total amount of enzyme is constant over time, so we can eliminate $\rm [E]$ using the relation $\rm [E] + [ES] = [E]_{tot}$, giving

$$(k_{-1} + k_2)[{\rm ES}] = k_1[{\rm S}]([{\rm E}]_{\rm tot} - [{\rm ES}])$$

Solving for $[{\rm ES}]$ gives

$$[{\rm ES}] = \frac{k_1[{\rm S}][{\rm E}]_{\rm tot}}{k_{-1} + k_2 + k_1[{\rm S}]}$$

Going back and inserting this into the reaction velocity gives

$$v = \frac{k_1 k_2 [{\rm S}][{\rm E}]_{\rm tot}}{k_{-1} + k_2 + k_1[{\rm S}]}.$$

This might look very messy, but remember that most of these are just constants, which we can recombine to simplify things. Defining $v_{\rm max} \equiv k_2[{\rm E}]_{\rm tot}$ gives

$$v = v_{\rm max} \frac{k_1 [{\rm S}]}{k_{-1} + k_2 + k_1[{\rm S}]}.$$

Next we divide by $k_1$ in the fraction and define $K_{m} = (k_{-1} + k_2)/k_1$, giving

$$v = v_{\rm max} \frac{[{\rm S}]}{K_m + [{\rm S}]}.$$

Note that $K_m > 0$, so that $[{\rm S}]/(K_m + [{\rm S}])<1$, meaning the reaction velocity can be in the range $[0, v_{\rm max}]$.

If $[S] = K_m$, we get $v = v_{\rm max}/2$, thus $K_m$ can be thought of as the point where the enzyme is half-saturated.

The model we have just derived is known as Michaelis-Menten kinetics. It works very well for a broad range of enzymes. The cases were it might not work that well are cases where we have several different substrates. Single substrate cases are usually assumed to follow Michaelis-Menten kinetics.

Also, take a note to compare the Michaelis-Menten kinetics with the hill equation. Note how similar these are. In fact, the Michaelis-Menten equation is a special case of the Hill equation where $n=1$, meaning we assume there is no cooperativity involved. That the Michaelis-Menten model gives the exact same expression as our derivation of the Hill equation isn't terribly surprising—but it does tell us that enzyme kinematics exhibit the same saturation behavior as the fractional binding we look at earlier.

### Enzyme Inhibition

Another class of molecules are the enzyme *inhibitors*. These are molecules that reduce, or completely shut of, the activity of enzymes. There are different functional ways for inhibitors to work, but the simplest mechanism is simply binding to enzyme more readily than the actual substrate, but then not actually converting to anything. Thus the enzymes waste their binding time on the inhibitors rather than the substrate. 

Enzyme inhibition is another important tool in the cell for regulating behavior, and many drugs are enzyme inhibitors. An example of this is penicillin, which inhibits an enzyme bacteria need for cell wall synthesis, eventually killing the bacteria as it is unable to maintain its own cell wall.

Extending the Michaelis-Menten model to include enzyme inhibitors is quite easy, we simply add the inhibitor as a new species $I$, and new reactions to the system where the inhibitor can bind to the enzyme. It then turns out we can make slightly different models, depending on where the inhibitor binds. These different models would behave differently in practice. When comparing to experiment, it turns out that the different models correspond to different inhibitors, and so in practice, different enzyme inhibitors are grouped into different classes. #### Law of Mass Action applied to ion channel gating

The most important thing we covered in this lecture was the law of mass action. We introduced this as a physical law govering chemical reactions. As mentioned however, the mathematical framework is a lot more general, and can be applied to physical processes as well. An important example of this, with respect to this course, is ion channel gating. A given *ion channel* (more on these soon), can be in either an *open* or a *closed* state. A single cell will a large number of these channels, and so we can create a variable that represents the *ratio* of open or closed channels, and denote these $O$ and $C$. Such that 

$$C + O = 1$$. 

To model how these ion channels open or close over time, we can now write this out as a "reaction" and apply the law of mass action

$${\rm C} \underset{\beta}{\overset{\alpha}\rightleftharpoons} {\rm O}$$


Exactly the mechanism for how the ion channel opens is not that important in this case, just that there are some rates for which channels open and close, typically called the *opening* and *closing* rates or simply *on*/*off* rates. Here we denote them $\alpha$ and $\beta$. We can then set up the ODE

$$\frac{{\rm d}O}{{\rm d}t} = \alpha C - \beta O.$$

Because we know that $O + C = 1$ we can write this out as a ODE of a single variable,

$$\frac{{\rm d}O}{{\rm d}t} = \alpha(1 - O) - \beta O = \alpha - (\alpha + \beta) O.$$

And we can solve this ODE numerically just like our earlier examples. What makes this system a bit more complex than the earlier examples we have shown, is that the on and off rates for ion channels, the $\alpha$ and $\beta$, are typically note static model parameters — but rather full ODE states that vary over time. But you will learn more about that later.
